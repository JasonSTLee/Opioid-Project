```python
import pandas as pd
import os
import glob
```

```python
file_path = '/Users/admin/Desktop/states/arcos-*-statewide-itemized.csv'
files = glob.glob(file_path)
```

```python
for file in files:
    df = pd.read_csv(file, usecols= ['TRANSACTION_DATE',
                                     'REPORTER_DEA_NO',
                                     'REPORTER_BUS_ACT',
                                     'REPORTER_NAME',
                                     'REPORTER_COUNTY',
                                     'REPORTER_STATE',
                                     'BUYER_DEA_NO',
                                     'BUYER_BUS_ACT',
                                     'BUYER_NAME',
                                     'BUYER_COUNTY',
                                     'BUYER_STATE',
                                     'DRUG_NAME',
                                     'Combined_Labeler_Name',
                                     'Reporter_family',
                                     'CALC_BASE_WT_IN_GM',
                                     'DOSAGE_UNIT'])
    
    state_abbreviation = file.split('-')[1]
    
    state_name = f'{state_abbreviation}.csv'
    df.to_csv(state_name, index= False)
    
    print(f"Ready: {state_abbreviation} as {state_name}")
```
